# Iris Dataset Exploration and Visualization

### Objective
Learn how to load, inspect, and visualize a dataset to understand data trends, distributions, and outliers using Python's standard data science ecosystem:
- **Pandas** for data loading and summary statistics
- **Matplotlib** and **Seaborn** for high-quality data visualizations

---

### Step 1: Import Required Libraries
We will load standard packages `pandas`, `numpy`, `matplotlib.pyplot`, and `seaborn`.

```python
import os
import urllib.request
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

print("Libraries successfully imported!")
```

### Step 2: Download and Load the Dataset
We will first check if `iris.csv` exists locally. If not, we will download it from Seaborn's public dataset repository and save it. Then, we will load it into a pandas DataFrame.

```python
DATASET_URL = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"
DATASET_PATH = "iris.csv"

# Download locally if not exists
if not os.path.exists(DATASET_PATH):
    print(f"Downloading dataset from: {DATASET_URL}...")
    urllib.request.urlretrieve(DATASET_URL, DATASET_PATH)
    print("Download complete!")
else:
    print("Local dataset copy found. Loading from disk...")

# Load using pandas
df = pd.read_csv(DATASET_PATH)
print(f"Loaded dataset successfully. Dimensions: {df.shape}")
```

### Step 3: Inspect the Dataset
Let's print the basic attributes of our DataFrame:
1. **Shape**: Total rows and columns.
2. **Column Names**: The headers representing our features and target.
3. **First 5 rows (`.head()`)**: Quick preview of the raw data values.

```python
# Print shape
print(f"Dataset Shape: {df.shape}\n")

# Print column names
print(f"Columns: {list(df.columns)}\n")

# Display first few rows
df.head()
```

### Step 4: Summary Statistics & Class Distribution
To understand our features better, we inspect:
- **`df.info()`**: Information about data types, non-null counts, and memory usage.
- **`df.describe()`**: Standard statistical summaries (mean, standard deviation, min, max, percentiles).
- **`df['species'].value_counts()`**: Verifying the frequency of each Iris class (setosa, versicolor, virginica).

```python
# Output dataset summary
print("=== Dataset Info ===")
df.info()

print("\n=== Descriptive Statistics ===")
df.describe()
```

```python
# Verify class balance
print("=== Species Distribution ===")
print(df['species'].value_counts())
```

### Step 5: Data Visualization
We will set up a premium, unified theme and palette. We will use a custom palette of modern colors: Indigo, Emerald, and Amber.

```python
# Setup style and aesthetics
sns.set_theme(style="whitegrid", context="talk")
plt.rcParams.update({
    'figure.autolayout': True,
    'font.family': 'sans-serif',
    'axes.titleweight': 'bold',
    'axes.labelweight': 'semibold',
    'figure.titlesize': 18,
    'axes.titlesize': 14,
    'axes.labelsize': 12
})

# Custom premium color palette
palette = {"setosa": "#4F46E5", "versicolor": "#10B981", "virginica": "#F59E0B"}
```

#### 1. Scatter Plots (Feature Relationships)
We'll visualize relationships between Sepal dimensions and Petal dimensions using scatter plots colored by species.

```python
# Create scatter plots
fig, axes = plt.subplots(1, 2, figsize=(16, 7))

# Sepal Dimensions Scatter
sns.scatterplot(
    data=df, 
    x="sepal_length", 
    y="sepal_width", 
    hue="species", 
    palette=palette,
    s=85, 
    alpha=0.85, 
    edgecolor="w", 
    linewidth=0.75,
    ax=axes[0]
)
axes[0].set_title("Sepal Length vs Sepal Width")
axes[0].set_xlabel("Sepal Length (cm)")
axes[0].set_ylabel("Sepal Width (cm)")
axes[0].legend(title="Species", frameon=True, facecolor="white", edgecolor="none", shadow=True)

# Petal Dimensions Scatter
sns.scatterplot(
    data=df, 
    x="petal_length", 
    y="petal_width", 
    hue="species", 
    palette=palette,
    s=85, 
    alpha=0.85, 
    edgecolor="w", 
    linewidth=0.75,
    ax=axes[1]
)
axes[1].set_title("Petal Length vs Petal Width")
axes[1].set_xlabel("Petal Length (cm)")
axes[1].set_ylabel("Petal Width (cm)")
axes[1].legend(title="Species", frameon=True, facecolor="white", edgecolor="none", shadow=True)

plt.suptitle("Feature Relationships: Sepal vs Petal Dimensions", y=1.02)
plt.show()
```

#### 2. Histograms (Value Distributions)
Histograms show how the measurements are spread. Adding Kernel Density Estimation (KDE) outlines allows us to see the shape of distributions clearly.

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
features = ["sepal_length", "sepal_width", "petal_length", "petal_width"]
feature_labels = ["Sepal Length (cm)", "Sepal Width (cm)", "Petal Length (cm)", "Petal Width (cm)"]

for i, (feature, label) in enumerate(zip(features, feature_labels)):
    row, col = divmod(i, 2)
    ax = axes[row, col]
    sns.histplot(
        data=df,
        x=feature,
        hue="species",
        kde=True,
        multiple="layer",
        palette=palette,
        alpha=0.45,
        element="bars",
        edgecolor="w",
        linewidth=0.5,
        ax=ax,
        legend=(i == 3)
    )
    ax.set_title(f"Distribution of {label.split(' (')[0]}")
    ax.set_xlabel(label)
    ax.set_ylabel("Count")
    if i == 3 and ax.get_legend() is not None:
        ax.get_legend().set_title("Species")

fig.suptitle("Value Distributions of Iris Features", y=1.01)
plt.show()
```

#### 3. Box Plots (Outlier Identification)
Box plots outline the median, quartiles, and range of distributions. Points appearing beyond the whisker lines represent statistical outliers, which we will highlight in red.

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

for i, (feature, label) in enumerate(zip(features, feature_labels)):
    row, col = divmod(i, 2)
    ax = axes[row, col]
    sns.boxplot(
        data=df,
        x="species",
        y=feature,
        palette=palette,
        hue="species",
        legend=False,
        width=0.6,
        linewidth=1.5,
        fliersize=6,
        flierprops={"markerfacecolor": "red", "markeredgecolor": "black", "marker": "o"},
        ax=ax
    )
    ax.set_title(f"{label.split(' (')[0]} by Species")
    ax.set_xlabel("Species")
    ax.set_ylabel(label)

fig.suptitle("Feature Box Plots by Species (Outliers in Red)", y=1.01)
plt.show()
```

---
### Step 6: Summary of Findings
Based on the plots generated above, we can draw some direct observations:
1. **Species Separability**: 
   - **Iris-setosa** is linearly separable from the other two species in both scatter plots, especially when plotting Petal Length vs Petal Width. It has much smaller and more clustered petals.
   - **Iris-versicolor** and **Iris-virginica** have some overlapping distributions but show clear boundaries, particularly in petal measurements.
2. **Feature Distributions**:
   - Petal measurements exhibit a highly bimodal shape due to setosa being very small and the other species being larger.
   - Sepal length and width show a normal-like distribution (more unimodal).
3. **Outliers**:
   - Looking at the box plots, **Iris-setosa** shows some minor outliers in **sepal width** and **petal length** (individual red points).
   - **Iris-versicolor** has a few minor outliers in **petal length**.
   - These outliers can be noted during preprocessing if we train machine learning models later.
